```python
df = pd.read_csv("./data/clear_data.csv", )
df.info()
```

```python
df_short = df.sample(1000)
```

```python
def jaccard(a, b):
    if not a and not b:
        return 1.0
    return len(a & b) / len(a | b)
```

```python
errors = []
scores = []
for _, row in df.iterrows():
    doc = nlp(row['letter_clean'])
    pred = set([i.text for i in doc.ents])    
    target = set( [i.strip() for i in row['target'].split(",")])
    score = jaccard(pred, target)
    scores.append(score)
    if target != pred:
        errors.append( [row, pred])
        print(f"Найдена ошибка: exp: {target}, but output {pred}",)
```
